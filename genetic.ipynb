```python
import random

def get_random(l):
    x = [random.randint(-100, 100) for _ in range(l)]
    return x
```

```python
class Person:
    #особь - вектор x = [x1, x2, x3, x4....]
    def __init__(self, x):
        self.lenght = len(x)
        self.x = get_random(len(x))
        
    #фитнесс функция, если решение правильное - возвращает 0
    def fitness(self, k):
        return (sum([k[i]*self.x[i] for i in range(self.lenght)]) - k[-1])

    def get_children(self, person2):
        x = [self.x[i] + person2.x[i]  for i in range(len(self.x))]
        return Person(x)
```

```python
coef_mutations = 0.45

class Population:
    def __init__(self, population):
        self.population = population
    
    def mutate(self):
        mutations_count = int(len(self.population) * coef_mutations) + 1 
      
        for i in range(1, mutations_count):
            elem = random.randint(0, len(self.population) - 1) #Выбор особи для мутации
            p = self.population.pop(elem)
            ex = p.x
            index = random.randint(0, p.lenght-1)
            ex.pop(index)
            ex.insert(index, randint(-100, 100))
            self.population.append(Person(ex))

    def cross(self, k):
        parent1 = random.sample(self.population, len(self.population))
        parent2 = random.sample(self.population, len(self.population))
        child = list()

        for i in range(0, len(parent1)):
            child.append(parent1[i].get_children(parent2[i]))

        # выбираем лучшую особь
        def check(a, b):
            if abs(a.fitness(k)) < abs(b.fitness(k)):
                return a
            else:
                return b

        self.population = list(map(check, self.population, child))
```

```python
#k1 * x1 + k2 * x2 + k3 * x3 = k4 (вид уравнения)
len_x = 3                 #количество х
len_k = len_x + 1         #количество коэффециентов
count_population = 4200   #количество особей в популяции
generation_count = 42     #кол-во поколений

k = [1,2,3,42] 
#Можно зарандомить коэффиценты, но возможно будет дольше искать
#k = get_random(len_k)

pops = Population([Person(get_random(len_x)) for _ in range(count_population)])

# cкрещиваем особей и мутируем
for i in range(1, generation_count):
    pops.cross(k)
    pops.mutate()

# отбираем особей
res = [i for i in pops.population if i.fitness(k) == 0]

#Вывод результатов
if len(res) != 0:
    str = f'Решения уравнения '
    str += f'{k[0]}x{0}'
    for i in range(1, len(k) - 1):
        str += f' + {k[i]}x{i}'
    str += f' = {k[-1]}'
    print(str)
    for i in res:
        str = ''
        for j in range(len(i.x)):
            str += f'x{j} = {i.x[j]} '
        print(str)
else:
    print('Упс, не повезло. Попробуйте еще раз')
```
